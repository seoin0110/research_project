```python
# English
!pip install jamo
from IPython.display import Audio
import time
import importlib
import yaml
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf

from tensorflow_tts.inference import AutoConfig
from tensorflow_tts.inference import TFAutoModel
from tensorflow_tts.inference import AutoProcessor

config = AutoConfig.from_pretrained("tensorspeech/tts-tacotron2-ljspeech-en")
import IPython.display as ipd

processor = AutoProcessor.from_pretrained("tensorspeech/tts-tacotron2-ljspeech-en")
tacotron2 = TFAutoModel.from_pretrained("tensorspeech/tts-tacotron2-ljspeech-en")
tacotron2.setup_window(win_front=6, win_back=6)
tacotron2.setup_maximum_iterations(3000)
melgan = TFAutoModel.from_pretrained("tensorspeech/tts-melgan-ljspeech-en")


input_text = "Isn't it Mr. Lee? Hello. You're coming to work early. I couldn't work because I was tired all day when I took the full subway. Me too. So I decided to give up breakfast and go to work 30 minutes early."


```

```python
input_text = "Isn't it Mr. Lee? Hello. You're coming to work early. I couldn't work because I was tired all day when I took the full subway. Me too. So I decided to give up breakfast and go to work 30 minutes early."
print("processor start")
input_ids = processor.text_to_sequence(input_text)
print("Input IDs Shape:",np.array(input_ids).shape)

print("tacotron2 start")
decoder_output, mel_outputs, stop_token_prediction, alignment_history = tacotron2.inference(
    tf.expand_dims(tf.convert_to_tensor(input_ids, dtype=tf.int32), 0),
    tf.convert_to_tensor([len(input_ids)], tf.int32),
    tf.convert_to_tensor([0], dtype=tf.int32)
)

print("Decoder Output Shape:", decoder_output.numpy().shape)
print("Mel Outputs Shape:", mel_outputs.numpy().shape)
print("Stop Token Prediction Shape:", stop_token_prediction.numpy().shape)
print("Alignment History Shape:", alignment_history.numpy().shape)

print("melgan start")
audio = melgan.inference(mel_outputs)[0, :, 0]
print("Audio Shape:", audio.numpy().shape)

audio = audio / tf.reduce_max(tf.abs(audio))
```

```python
Audio(data=np.array(audio), rate=22050)
```
