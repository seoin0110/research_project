```python
# Input 데이터를 지정된 형식에 맞게 변환합니다.
def prepare_input(input_ids):
  return (tf.expand_dims(tf.convert_to_tensor(input_ids, dtype=tf.int32), 0),
          tf.convert_to_tensor([len(input_ids)], tf.int32),
          tf.convert_to_tensor([0], dtype=tf.int32))

# TTS 모델 Inference.
def infer(input_text):
  processor = AutoProcessor.from_pretrained(pretrained_path="test/files/kss_mapper.json")
  input_ids = processor.text_to_sequence(input_text)
  interpreter.resize_tensor_input(input_details[0]['index'], [1, len(input_ids)])
  interpreter.allocate_tensors()
  input_data = prepare_input(input_ids)

  for i, detail in enumerate(input_details):
    input_shape = detail['shape']
    interpreter.set_tensor(detail['index'], input_data[i])

  interpreter.invoke()

  return (interpreter.get_tensor(output_details[0]['index']),
          interpreter.get_tensor(output_details[1]['index']))
```

```python
#tts setting with gpu

import time

import yaml
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow_tts.inference import AutoConfig, TFAutoModel, AutoProcessor
from IPython.display import Audio

# TensorFlow가 GPU를 사용하는지 확인하기 위한 설정
tf.debugging.set_log_device_placement(True)

# 프로세서 로드
tts_processor = AutoProcessor.from_pretrained("tensorspeech/tts-tacotron2-ljspeech-en")
# 모델 로드
tacotron2 = TFAutoModel.from_pretrained("tensorspeech/tts-tacotron2-ljspeech-en")
# 설정
tacotron2.setup_window(win_front=6, win_back=6)
tacotron2.setup_maximum_iterations(3000)
melgan = TFAutoModel.from_pretrained("tensorspeech/tts-melgan-ljspeech-en")
```

```python
# translate setting

import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import sacrebleu
from nltk.translate.chrf_score import sentence_chrf
from tqdm import tqdm
import time

# Set device to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load model and tokenizer, then move the model to the GPU
def load_model_and_tokenizer(model_name):
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(device)
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    return model, tokenizer

# Translation function that uses the GPU
def translate(model, tokenizer, text):
    model.eval()
    encoded = tokenizer(text, return_tensors="pt", padding=True, truncation=True).to(device)

    with torch.no_grad():
        start_time = time.time()
        generated_tokens = model.generate(**encoded)
        latency = time.time() - start_time

    translation = tokenizer.decode(generated_tokens[0], skip_special_tokens=True)
    return translation, latency

# Load the model and tokenizer and move them to the GPU
translate_model_name = 'Helsinki-NLP/opus-mt-ko-en'
translate_model, tokenizer = load_model_and_tokenizer(translate_model_name)
```

```python
# 번역 + tts

input_text = "나는 지금 공부 중이야."

translation, latency = translate(translate_model, tokenizer, input_text)

# tts
start_time = time.time()
input_ids = tts_processor.text_to_sequence(translation)
print("tts pre_processing : ", time.time() - start_time)
# 추론을 GPU에서 실행
with tf.device('/GPU:0'):
    start_time = time.time()
    decoder_output, mel_outputs, stop_token_prediction, alignment_history = tacotron2.inference(
        tf.expand_dims(tf.convert_to_tensor(input_ids, dtype=tf.int32), 0),
        tf.convert_to_tensor([len(input_ids)], tf.int32),
        tf.convert_to_tensor([0], dtype=tf.int32)
    )
    print("tts inferencing : ", time.time() - start_time)
start_time = time.time()
audio = melgan.inference(mel_outputs)[0, :, 0]
print("tts melgan inferencing : ", time.time() - start_time)
audio = audio / tf.reduce_max(tf.abs(audio))

print(f"번역 Latency: {latency:.4f} seconds")

Audio(data=np.array(audio), rate=22050)
```

```python
# stt setting
!pip install h5py
!pip install typing-extensions
!pip install wheel
!pip install -r drive/MyDrive/연구프로젝트/requirements.txt
!pip install faster-whisper
!pip install jiwer Levenshtein tdqm
!pip install nvidia-cublas-cu12 nvidia-cudnn-cu12

!export LD_LIBRARY_PATH=`python3 -c 'import os; import nvidia.cublas.lib; import nvidia.cudnn.lib; print(os.path.dirname(nvidia.cublas.lib.__file__) + ":" + os.path.dirname(nvidia.cudnn.lib.__file__))'`

import ctranslate2
import librosa
import transformers
import os
import numpy as np
import jiwer
import json
import time
import Levenshtein as Lev
from tqdm import tqdm
import gc
import torch

os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
model_name = "drive/MyDrive/연구프로젝트/whisper-base-ct2"
model_type = "openai/whisper-base"
whisper_processor = transformers.WhisperProcessor.from_pretrained(model_type)
whisper_model = ctranslate2.models.Whisper(model_name, device="cuda")
prompt = whisper_processor.tokenizer.convert_tokens_to_ids(
    [
        "<|startoftranscript|>",
        "<|ko|>",
        "<|transcribe|>",
        "<|notimestamps|>",  # Remove this token to generate timestamps.
    ]
)

def apply_whisper(file_path, model_name, model_type, device="cuda"):
    audio, _ = librosa.load(file_path, sr=16000, mono=True)
    # Compute the features of the first 30 seconds of audio.

    inputs = whisper_processor(audio, return_tensors="np", sampling_rate=16000)
    features = ctranslate2.StorageView.from_array(inputs.input_features)

    # Run generation for the 30-second window.
    start_time = time.time()
    results = whisper_model.generate(features, [prompt])
    latency = time.time() - start_time
    transcription = whisper_processor.decode(results[0].sequences_ids[0])
    return transcription, latency
```

```python
# 100개 파일 stt + 번역 + tts

def do_one_process(model_name, model_type):
    total_latency = 0
    count = 0
    for root, dirs, files in os.walk('drive/MyDrive/연구프로젝트/data/voice'):
        for file in tqdm(files):
            if file.endswith(".wav"):
                count += 1
                voice_file = os.path.join(root, file)
                input_text, whisper_latency = apply_whisper(voice_file, model_name, model_type)
                translation, latency = translate(translate_model, tokenizer, input_text)

                # tts

                input_ids = tts_processor.text_to_sequence(translation)

                # 추론을 GPU에서 실행
                tts_start = time.time()
                with tf.device('/GPU:0'):
                    decoder_output, mel_outputs, stop_token_prediction, alignment_history = tacotron2.inference(
                        tf.expand_dims(tf.convert_to_tensor(input_ids, dtype=tf.int32), 0),
                        tf.convert_to_tensor([len(input_ids)], tf.int32),
                        tf.convert_to_tensor([0], dtype=tf.int32)
                    )

                audio = melgan.inference(mel_outputs)[0, :, 0]
                total_latency += whisper_latency + latency + (time.time() - tts_start)
                audio = audio / tf.reduce_max(tf.abs(audio))
    elapsed_time = total_latency
    average_time = total_latency / count
    print(f"Total stt+translate+tts elapsed time: {elapsed_time:.4f} seconds")
    print(f"average elapsed time: {average_time:.4f} seconds")

do_one_process(model_name, model_type)
```
